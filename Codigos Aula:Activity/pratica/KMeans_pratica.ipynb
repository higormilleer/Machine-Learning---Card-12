```python
from numpy import random, array

# Cria dados de pessoas agrupadas com renda e idade para N pessoas em k grupos
def createClusteredData(N, k):
    random.seed(10)  # Semente para gerar resultados reproduzíveis
    pointsPerCluster = float(N) / k  # Calcula o número de pontos por grupo
    X = []  # Lista para armazenar os pontos

    # Gera k clusters
    for _ in range(k):
        incomeCentroid = random.uniform(20000.0, 200000.0)  # Centroide aleatório para a renda
        ageCentroid = random.uniform(20.0, 70.0)  # Centroide aleatório para a idade

        # Cria os pontos para cada cluster
        for _ in range(int(pointsPerCluster)):
            # Adiciona pontos distribuídos normalmente em torno dos centroides
            X.append([random.normal(incomeCentroid, 10000.0), random.normal(ageCentroid, 2.0)])

    # Converte a lista X para um array numpy
    X = array(X)
    return X
```

```python
%matplotlib inline

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from numpy import random

data = createClusteredData(100, 4) #100 pessoas em 5 grupos direntes

model = KMeans(n_clusters=5)

# Note I'm scaling the data to normalize it! Important for good results.
model = model.fit(scale(data))

# We can look at the clusters each data point was assigned to
print(model.labels_)

# And we'll visualize it:
plt.figure(figsize=(8, 6)) #criando o grid
plt.scatter(x=data[:,0], y=data[:,1], c=model.labels_.astype(float)) #atribui cor aos rotulos anteriormente criados
plt.show()
```
